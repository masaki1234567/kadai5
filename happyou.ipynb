## py4cytoscapeをインポート

Cytoscapeをあらかじめ開いておき、py4cytoscapeを使いCytoscapeと接続できているか確認します。



```python
import py4cytoscape as p4c
p4c.cytoscape_ping()
p4c.cytoscape_version_info()
```

## ネットワーク（sif）、ノード属性（tsv）をCytoscapeにインポート

ノード属性テーブルのIDは型がint64で数字として扱われているので、文字列に変更させないといけません。

```python
import pandas as pd
#ネットワークファイルを必ず先にインポートします
p4c.import_network_from_file('7network.sif')
#ノード属性ファイルをインポートします
node7 = pd.read_csv('7node.tsv', sep='\t', encoding='utf-8')
#IDを文字列に変更させます（int64 → object）
node7['ID'] = node7['ID'].astype(str)
#pandasで読み込ませた属性tsvファイルをCytoscapeに渡す
p4c.load_table_data(node7, data_key_column='ID')
```

## py4cytoscapeを使いグラフの見た目を調整する

```python
#現時点のグラフ
p4c.notebook_export_show_image()
```

```python
#新しいスタイルを作成します
style_name = "myStyle"
p4c.create_visual_style(style_name)
p4c.set_visual_style(style_name)
```

```python
#所属のユニークな値を取得、ｄはdiscretemappingの意味です
from py4cytoscape import gen_node_color_map
p4c.set_node_color_mapping(**gen_node_color_map('所属', mapping_type='d', style_name=style_name))
p4c.notebook_export_show_image()
```

```python
#ボーダーの色をコミュニティごとに設定します
p4c.set_node_border_color_mapping(**gen_node_color_map('コミュニティ', mapping_type='d', style_name=style_name))
p4c.notebook_export_show_image()
```

```python
#ボーダーの幅をすべてのノードに設定します
p4c.set_node_border_width_default(new_width=5, style_name=style_name)
p4c.notebook_export_show_image()
```

```python
#ノードの高さ、幅を変更
p4c.set_node_height_default(new_height=15, style_name=style_name)
p4c.set_node_width_default(new_width=35, style_name=style_name)
p4c.notebook_export_show_image()
```

```python
#セッションを保存します
p4c.save_session('session7.cys')
```
